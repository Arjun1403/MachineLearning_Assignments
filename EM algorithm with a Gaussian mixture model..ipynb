# EM Algorithm

```python
def iris_expectation_maximization(df,parameters):
    
    try:
        for i in range(20):
            count=0
            means=parameters['cluster_feature_mean']
            probable=[]
            covariance=parameters['cluster_cov_matrix']
            for j in range(clusters):
                probable.append(multivariate_normal.pdf(x=iris_df,mean=means[j],cov=covariance[j]))
    #             probable.append(norm_pdf_multivariate(x=iris_df, mu=means[j], sigma=covariance[j]))


            probable = np.array(probable)
            print(f'likelyhood length ={len(probable.flatten())}')
            predictions = np.argmax(probable, axis=0)
            build=[]
            weights=parameters['cluster_weights']
            
            for max_cluster in range(clusters):
                build.append((probable[max_cluster] * weights[max_cluster]) / (np.sum([probable[i] * weights[i] for i in range(clusters)], axis=0)))
                print(f'before={parameters["cluster_feature_mean"]}')
                parameters['cluster_feature_mean'][max_cluster]= np.sum(build[max_cluster].reshape(iris_df.shape[0],1) * iris_df, axis=0) / (np.sum(build[max_cluster]))
                print(f' after ={parameters["cluster_feature_mean"]}')
                parameters['cluster_cov_matrix'][max_cluster]=np.dot((build[max_cluster].reshape(iris_df.shape[0],1) * (iris_df - means[max_cluster])).T, (iris_df - means[max_cluster])) / (np.sum(build[max_cluster]))
                parameters['cluster_weights'][max_cluster]=np.mean(build[max_cluster])
#                 predicted_cluster.count(max_cluster)/iris_df.shape[0]

            
    #         print(parameters)
            print(set(predictions))
    except Exception as E:
        print(E)
        return predictions
        

    return predictions
            
            
    
```

# Predicitons on using EM algorithm

```python
# clusters =2
y_2actual=iris_expectation_maximization(iris_df,parameters)
fig=px.scatter(iris_df,x=0,y=1,color=list(map(lambda x:x+1,y_2actual)))
fig.show()
```

# Accuracy calculations for clusters 2,3 and 4

```python
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

def kmeans(clusters,X):
    model = KMeans(n_clusters=clusters)
    clusters_labels = model.fit_predict(X)
    return clusters_labels
    
y_predict_3means=kmeans(3,iris_df)
y_predict_2means=kmeans(2,iris_df)
y_predict_4means=kmeans(4,iris_df)
```

```python
# 2 clusters  accuracy with kmeans
accuracy_score(y_pred=y_2actual,y_true=y_predict_2means)
```

```python
# 3 clusters accuracy with kmeans
accuracy_score(y_pred=y_3actual,y_true=y_predict_3means)
```

```python
# 4 clusters accuracy with kmeans
accuracy_score(y_pred=y_4actual,y_true=y_predict_4means)
```
